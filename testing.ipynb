```python

from fetch_article_html import get_article_html
from extract_references import extract_external_links, filter_links_for_checking
```

```python


def test_functions(test_title="Python_(programming_language)"):
    """Test both functions with a sample Wikipedia article."""
    
    # Step 1: Fetch the article HTML
    html_content = get_article_html(test_title)
    
    # Step 2: Extract external links
    external_links = extract_external_links(html_content)
    
     # Step 3: Filter links for checking (remove archives, group duplicates)
    links_to_check, archive_groups = filter_links_for_checking(external_links)
    return external_links,links_to_check, archive_groups, html_content
```
